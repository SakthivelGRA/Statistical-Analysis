## Treating the outliers in "Total_salary" column using IQR method

```python
Q1 = df1['Total_salary'].quantile(0.25)
Q3 = df1['Total_salary'].quantile(0.75)
IQR = Q3 - Q1
```

```python
Q1,Q3,IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound,upper_bound
```

```python
outliers = df1[(df1['Total_salary'] < lower_bound) | (df1['Total_salary'] > upper_bound)]
```

```python
outliers
```

```python
#Removing outliers
df_no_outliers = df1[(df1['Total_salary'] >= lower_bound) & (df1['Total_salary'] <= upper_bound)]
```

```python
df_no_outliers
```

```python
# Treating the outliers
df1_tre = df1.copy()
df1_tre.loc[df1_tre['Total_salary'] < lower_bound,'Total_salary'] = lower_bound
df1_tre.loc[df1_tre['Total_salary'] > upper_bound,'Total_salary'] = upper_bound

```

```python
df1_tre
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x=df1['Total_salary'])
plt.title('Boxplot (Before Outlier Treatment)')

plt.subplot(1, 3, 2)
sns.boxplot(x=df_no_outliers['Total_salary'])
plt.title('Boxplot (After Removing Outliers)')

plt.subplot(1, 3, 3)
sns.boxplot(x=df1_tre['Total_salary'])
plt.title('Boxplot (After Treating Outliers)')

plt.tight_layout()
plt.show()
```

# Likelihood of a salaried person buying a Sedan

```python
salary_bins = [0, 50000, 100000, 150000, float('inf')]
salary_labels = ['0-50k', '50k-100k', '100k-150k', '150k+']

df1['SalaryRange'] = pd.cut(df1['Salary'], bins=salary_bins, labels=salary_labels, right=False)
```

```python
sedan_data = df1.groupby('SalaryRange')['Make'].apply(lambda x: (x == 'Sedan').mean()).reset_index()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='SalaryRange', y='Make', data=sedan_data)
plt.title('Likelihood of  a salaried person buying a Sedan')
plt.xlabel('Salary Range')
plt.ylabel('Likelihood')
plt.show()
```
